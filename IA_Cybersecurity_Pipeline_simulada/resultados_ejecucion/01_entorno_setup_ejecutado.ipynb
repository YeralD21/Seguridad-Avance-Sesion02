## 2.4. Verificación del entorno
Este script verifica que todas las bibliotecas críticas se hayan instalado correctamente y muestra sus versiones.

````python
import sys
import importlib

# Listing 2.3: Script de verificación del entorno (Ajustado para funcionalidad)
REQUIRED = [
    ("numpy", "numpy"), 
    ("pandas", "pandas"), 
    ("scikit-learn", "sklearn"), 
    ("tensorflow", "tensorflow"),
    ("matplotlib", "matplotlib"), 
    ("seaborn", "seaborn"), 
    ("pefile", "pefile"), 
    ("requests", "requests"),
    ("imbalanced-learn", "imblearn"),
    ("shap", "shap")
]

print(f"Python {sys.version}\n{'='*50}")

for pkg_name, import_name in REQUIRED:
    try:
        mod = importlib.import_module(import_name)
        # Intentar obtener la versión, si no existe poner OK
        version = getattr(mod, "__version__", "OK")
        print(f"[OK] {pkg_name:<20} {version}")
    except ImportError:
        print(f"[FALTA] {pkg_name}")
        print(f"      -> Instalar con: !pip install {pkg_name}")
````

## 2.5. Estructura de Trabajo
Como paso final, aseguramos que existan las carpetas para almacenar los datos y los modelos generados en los siguientes capítulos.

````python
import os

def create_folders():
    folders = ['data', 'models']
    for folder in folders:
        if not os.path.exists(folder):
            os.makedirs(folder)
            print(f"[OK] Directorio creado: {folder}")
        else:
            print(f"[INFO] El directorio {folder} ya existe.")

create_folders()
````
