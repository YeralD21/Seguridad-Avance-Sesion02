## Validación del Entorno
Ejecutamos un script que verifica la disponibilidad de todas las bibliotecas críticas
y reporta sus versiones para confirmar compatibilidad.

````python
import sys
import importlib

# Verificación de paquetes instalados
REQUIRED_PACKAGES = [
    ("numpy", "numpy"), 
    ("pandas", "pandas"), 
    ("scikit-learn", "sklearn"), 
    ("tensorflow", "tensorflow"),
    ("matplotlib", "matplotlib"), 
    ("seaborn", "seaborn"), 
    ("pefile", "pefile"), 
    ("requests", "requests"),
    ("imbalanced-learn", "imblearn"),
    ("shap", "shap")
]

print(f"Sistema: Python {sys.version}\n{'─'*60}")

missing = []
for pkg_name, import_name in REQUIRED_PACKAGES:
    try:
        mod = importlib.import_module(import_name)
        version = getattr(mod, "__version__", "instalado")
        print(f"[✓] {pkg_name:<20} {version}")
    except ImportError:
        missing.append(pkg_name)
        print(f"[✗] {pkg_name}")

print('─'*60)
if missing:
    print(f"Faltantes: {', '.join(missing)}")
    print("Ejecutar: !pip install " + " ".join(missing))
else:
    print("✓ Todas las dependencias están disponibles.")
````

## Inicialización de Directorios
Creamos la estructura de carpetas necesaria para almacenar datos, modelos entrenados
y resultados de análisis a lo largo del pipeline.

````python
import os

def crear_estructura_directorios(directorios=['data', 'models', 'resultados']):
    """
    Crea los directorios necesarios para el pipeline.
    """
    for directorio in directorios:
        if not os.path.exists(directorio):
            os.makedirs(directorio)
            print(f"[+] Directorio creado: {directorio}")
        else:
            print(f"[→] Directorio existente: {directorio}")

crear_estructura_directorios()
print("\n✓ Estructura de directorios lista.")
````
